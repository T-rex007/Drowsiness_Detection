```python
def eye_aspect_ratio(eye):
    A = eye.dist.euclidean(eye[1], eye[5])
    B = eye.dist.euclidean(eye[2], eye[4])
    
    C = dist.euclidean(eye[0], eye[3])
    eye = (A + B) / (2.0 * C)
    return eye

    
```
